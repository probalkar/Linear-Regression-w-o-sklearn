```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.genfromtxt("data.csv", delimiter=",")
X = data[:, 0].reshape(-1,1)
ones  = np.ones([X.shape[0], 1])
X = np.concatenate([ones, X],1)
y = data[:, 1].reshape(-1,1)
```

```python
plt.scatter(data[:,0].reshape(-1,1),y)
```

```python
alpha = 0.0001
iters = 1000

theta = np.array([[1.0,1.0]])
```

```python
def computeCost(X,y,theta):
    inner = np.power(((X @ theta.T)-y),2)
    return np.sum(inner)/(2*len(X))
```

```python
computeCost(X,y,theta)
```

```python
def gradientDescent(alpha, iters, X, y, theta):
    for i in range(iters):
        theta -= (alpha/len(X)) * np.sum((X @ theta.T - y), axis=0)
        cost = computeCost(X,y,theta)
        if i%10 == 0:
            print(cost)

    return (theta, cost)
```

```python
g, cost = gradientDescent(alpha, iters, X, y, theta)
print(g, cost)
```

```python
plt.scatter(data[:,0].reshape(-1,1), y)
axes = plt.gca()
x_vals = np.array(axes.get_xlim())
y_vals = g[0][0] + g[0][1]*x_vals
plt.plot(x_vals, y_vals, '--', c='r')
```
